```
import pandas as pd
import matplotlib
import numpy as np
import OpenEXR
import Imath
import matplotlib.pyplot as plt
import glob
import os
import shutil
import random

from google.colab import drive
drive.mount('/content/gdrive')
```

```
DATA_PATH = 'gdrive/My Drive/NN Capstone/Train_Input/'
TEST_PATH = 'gdrive/My Drive/NN Capstone/Test Data/'
```

```
input_data = [str(i).zfill(2) for i in range(50)]
test_samples = random.sample(input_data, 10)
print(test_samples)
```

```
def splitTestData():
  for view in glob.iglob(DATA_PATH + 'species_2/specimen_*/'):
   # print(view)
    for i in test_samples:
      #print(view + 'View_' + i + '_n.npy')
      head,tail = os.path.split(view)
      subhead, subtail = os.path.split(head)
      specimen = subtail
      subhead, subtail = os.path.split(subhead)
      species = subtail
      #print(specimen,species)
      #print(TEST_PATH + species + '/' + specimen + '/' +  'View_' + i + '_n.npy')
      shutil.move(view + 'View_' + i + '_n.npy', TEST_PATH + species + '/' + specimen + '/' +  'View_' + i + '_n.npy')
      shutil.move(view + 'View_' + i + '_d.npy', TEST_PATH + species + '/' + specimen + '/' +  'View_' + i + '_d.npy')
    
splitTestData()
```

```
def splitTestOutData():
  for view in glob.iglob(DATA_PATH + 'species_*/specimen_*/'):
   # print(view)
    head,tail = os.path.split(view)
    subhead, subtail = os.path.split(head)
    specimen = subtail
    subhead, subtail = os.path.split(subhead)
    species = subtail
    print(view + specimen +'.npy')
    #print(specimen,species)
    print(TEST_PATH + species + '/' + specimen + '/'  + specimen + '.npy')
    shutil.copyfile(view + specimen +'.npy', TEST_PATH + species + '/' + specimen + '/'  + specimen + '.npy')
    
splitTestOutData()
```
